## Построение модели:

В этом разделе мы создадим функцию, которая добавит новые признаки в датафрейм, чтобы улучшить качество модели. Это нужно для того, чтобы модель могла лучше учитывать временные зависимости и сезонные колебания, которые важны для прогнозирования заказов такси.

Мы добавим следующие признаки:

1. **Месяц, день, день недели, час** — эти признаки помогут модели учитывать сезонность и различные циклы в данных, например, более высокое количество заказов в определённые дни недели или часы.
2. **Лаги (shift)** — создадим признаки, которые будут отражать значения заказов с отставанием, например, за 1, 2, 3 и более часов. Это поможет модели учитывать, как предыдущие данные влияют на текущие.
3. **Скользящее среднее** — добавим признак, который покажет среднее количество заказов за определённый период времени, чтобы модель могла сгладить случайные колебания и выявить общие тренды.

Эти новые признаки позволят модели лучше понимать временные зависимости в данных, что должно повысить точность её прогнозов.

```python
# Создадим функцию, которая добавляет признаки для обучения модели 
def make_features(df, max_lag, rolling_mean_size):
    """
    Функция получает на вход оригинальный датафрейм и возвращает датафрейм с новыми признаками
    
    :param data (pd.DataFrame): Оригинальный датафрейм 
    :param max_lag(int): Максимальное предыдущее значение временного ряда
    :param rolling_mean_size(int): Значение ширины скользящего окна
    :return(pd.DataFrame): Датафрейм с новыми признаками 
    """
    data_copy = df.copy(deep=True)
    data_copy['month'] = data_copy.index.month
    data_copy['day'] = data_copy.index.day
    data_copy['dayofweek'] = data_copy.index.dayofweek
    data_copy['hour'] = data_copy.index.hour

    for lag in range(1, max_lag + 1):
        data_copy['lag_{}'.format(lag)] = data_copy['num_orders'].shift(lag)

    data_copy['rolling_mean'] = data_copy['num_orders'].shift().rolling(rolling_mean_size).mean()
    data_copy.dropna(inplace=True)

    return data_copy
```

Создадим датасет для обучения моделей:

```python
# Создадим датасет для обучения моделей
data_ml = make_features(data_taxi, 180, 180)
display(f'Размер оригинального датасета: {data_taxi.shape}')
display(f'Размер обновленного датасета: {data_ml.shape}')
```

Проведем разбиение на обучающую и тестовую выборки:

```python
# Подготовим обучающую и тестовую выборки
RANDOM_STATE = 42
TEST_SIZE = 0.1

# Разбиваем признаки и таргет
X = data_ml.drop('num_orders', axis=1)
y = data_ml['num_orders']

# Разбиваем выборку на тренировочную и тестовую
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    shuffle=False,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE)

# Проверим корректность разбивки и утечку в тестовой выборке
print('---')
print('Features')
print('---')
print(f'Временной промежуток обучающей выборки: {X_train.index.min()}, {X_train.index.max()}')
print(f'Временной промежуток тестовой выборки: {X_test.index.min()}, {X_test.index.max()}')
print('---')
print('Target')
print('---')
print(f'Временной промежуток обучающей выборки: {y_train.index.min()}, {y_train.index.max()}')
print(f'Временной промежуток тестовой выборки: {y_test.index.min()}, {y_test.index.max()}')
```

### LinearRegression

```python
# Создадим пустой список для моделей, RMSE на CV и RMSE на обучающей выборке
models = []
```

```python
%%time

# Обучение модели
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Кросс-валидация
cv = TimeSeriesSplit(n_splits=5)

# Проводим кросс-валидацию с использованием RMSE
scores = cross_val_score(
    lin_reg,
    X_train,
    y_train,
    cv=cv,
    n_jobs=-1,
    scoring='neg_root_mean_squared_error'  # Оценка по RMSE
)

# Среднее значение RMSE из кросс-валидации
rmse_cv_lin_reg = -scores.mean()
print(f'RMSE LinearRegression на кросс-валидации: {rmse_cv_lin_reg}')

# Оценка модели на тренировочной выборке
y_pred_lin_reg = lin_reg.predict(X_train)

# Вычисление RMSE на тренировочной выборке
rmse_train_lin_reg = np.sqrt(np.mean((y_train - y_pred_lin_reg) ** 2))
print(f'RMSE LinearRegression на тренировочной выборке: {rmse_train_lin_reg}')
print('---')

# Добавляем модель в список
models.append([lin_reg, rmse_cv_lin_reg, rmse_train_lin_reg])
```

### RandomForestRegressor

```python
%%time

# Инициализация модели случайного леса
rf_model = RandomForestRegressor(random_state=RANDOM_STATE)

# Параметры для подбора с помощью RandomizedSearchCV
rf_params = {
    'n_estimators': range(50, 100, 10),
    'max_depth': range(5, 20, 5),
}

# Инициализация RandomizedSearchCV
rf_grid = RandomizedSearchCV(
    rf_model,
    rf_params,
    scoring='neg_root_mean_squared_error',  # Используем метрику RMSE
    n_jobs=-1,
    n_iter=10,
    cv=cv,
    verbose=5,
    random_state=RANDOM_STATE
)

# Обучение модели с подбором гиперпараметров
rf_grid.fit(X_train, y_train)

# Получение лучшей модели
rf = rf_grid.best_estimator_

# Расчёт RMSE из результатов кросс-валидации
rmse_cv_rf = -rf_grid.best_score_  # Так как оценка в RandomizedSearchCV возвращается как отрицательное значение
print(f'Лучшая модель: \n {rf}')
print(f'RMSE RandomForestRegressor на кросс-валидации: {rmse_cv_rf}')

# Оценка модели на тренировочной выборке
y_pred_rf = rf.predict(X_train)

# Вычисление RMSE на тренировочной выборке
rmse_train_rf = np.sqrt(np.mean((y_train - y_pred_rf) ** 2))
print(f'RMSE RandomForestRegressor на тренировочной выборке: {rmse_train_rf}')
print('---')

# Добавление модели в список
models.append([rf, rmse_cv_rf, rmse_train_rf])
```

### CatBoostRegressor

```python
%%time

# Инициализация модели CatBoostRegressor
catboost_model = CatBoostRegressor(random_state=RANDOM_STATE, verbose=False)

# Обучение модели
catboost_model.fit(X_train, y_train)

# Оценка модели с помощью кросс-валидации
catboost_scores = cross_val_score(
    catboost_model,
    X_train,
    y_train,
    cv=cv,
    n_jobs=-1,
    scoring='neg_root_mean_squared_error'
)

# Расчёт RMSE на кросс-валидации
rmse_cv_catboost = -catboost_scores.mean()
print(f'Лучшая модель: \n {catboost_model}')
print(f'RMSE CatBoostRegressor на кросс-валидации: {rmse_cv_catboost}')

# Оценка модели на тренировочной выборке
y_pred_catboost = catboost_model.predict(X_train)

# Вычисление RMSE на тренировочной выборке
rmse_train_catboost = np.sqrt(np.mean((y_train - y_pred_catboost) ** 2))
print(f'RMSE CatBoostRegressor на тренировочной выборке: {rmse_train_catboost}')
print('---')

# Добавление модели в список
models.append([catboost_model, rmse_cv_catboost, rmse_train_catboost])
```

## Тестирование лучшей модели

```python
# Предсказания модели
y_pred = lin_reg.predict(X_test)

# Рассчитываем MSE
mse = mean_squared_error(y_test, y_pred)

# Извлекаем корень для получения RMSE
best_rmse = np.sqrt(mse)

# Выводим результат
print(f'RMSE лучшей модели на тестовой выборке: {best_rmse}')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Визуализируем предсказания и истинные значения
def display_result(target, pred, rmse):
    """
    Функция переводит правильные ответы в DataFrame, добавляет предсказания и строит два графика на единой плоскости.
    
    :param target(np.array / pd.Series): y_true, правильные ответы
    :param pred(np.array / pd.Series): y_pred, предсказанные ответы
    :param rmse: Лучшая полученная метрика качества
    :return: График
    """
    # Создаем DataFrame из истинных значений и предсказаний
    result = target.to_frame().reset_index()
    result['prediction'] = pd.Series(pred)
    result.set_index('datetime', inplace=True)
    
    # Создаем график
    plt.figure(figsize=(12, 6))

    # График истинных значений (первый столбец DataFrame)
    plt.plot(result.index, result[result.columns[0]], label='Истинные значения', color='blue')

    # График предсказанных значений
    plt.plot(result.index, result['prediction'], label='Предсказанные значения', color='orange')

    # Названия осей и графика
    plt.xlabel('Дата/Время')  # Название оси X
    plt.ylabel('Значение')  # Название оси Y
    plt.title(f'Сравнение истинных и предсказанных значений\nRMSE = {rmse}')  # Название графика

    # Легенда
    plt.legend()

    # Отображаем график
    plt.show()
```

```python
display_result(y_test, y_pred, best_rmse)
```

**Промежуточный вывод**: на тестовой выборке RMSE лучшей модели составил 35.45. Мы также построили график, сравнивающий истинные и предсказанные значения, и результат показал, что модель достаточно хорошо справляется с задачей прогнозирования.
